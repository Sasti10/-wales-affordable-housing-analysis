```python
][[import pandas as pd

# Load the file
df = pd.read_csv('/content/additional-affordable-housing-provision-by-location-provider-funding-and-method-of-delivery-v1-en-gb.csv')

# First look
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Check missing values in each column
print("Missing values per column:")
print(df.isnull().sum())

print("\nUnique years in data:")
print(df['Period'].unique())

print("\nUnique locations:")
print(df['Location'].unique())

print("\nUnique providers:")
print(df['Provider'].unique())
```

```python
# Keep only useful columns
df_clean = df[[
    'Data values_sort',
    'Period',
    'Location',
    'Provider',
    'Funding'
]].copy()

# Rename to simple names
df_clean.columns = ['Units', 'Year', 'Location', 'Provider', 'Funding']

# Check the result
print("Clean data shape:", df_clean.shape)
print("\nMissing values now:")
print(df_clean.isnull().sum())

df_clean.head(10)
```

```python
# Export clean data to CSV
df_clean.to_csv('wales_affordable_housing_clean.csv', index=False)

print("File saved successfully!")
print("Rows:", len(df_clean))
print("Columns:", df_clean.columns.tolist())
```

```python
# Which location delivered most housing overall?
print("Top 10 locations by total units:")
print(df_clean[df_clean['Location'] != 'Wales']
      .groupby('Location')['Units']
      .sum()
      .sort_values(ascending=False)
      .head(10))

# Which year had most delivery?
print("\nTotal units by year:")
print(df_clean[df_clean['Location'] == 'Wales']
      .groupby('Year')['Units']
      .sum())
```

```python
# Which provider delivers more?
print("Units by Provider type (Wales total only):")
print(df_clean[(df_clean['Location'] == 'Wales') &
               (df_clean['Funding'] == 'Total')]
      .groupby('Provider')['Units']
      .sum()
      .sort_values(ascending=False))

# Which funding type is more common?
print("\nUnits by Funding type:")
print(df_clean[(df_clean['Location'] == 'Wales') &
               (df_clean['Provider'].isin(['Local Authorities',
                'Registered Social Landlords']))]
      .groupby('Funding')['Units']
      .sum()
      .sort_values(ascending=False))
```
